```python
#Program 1: Even Number Skipper
number = 0
while number < 20:
  number += 1
  if number % 2 == 0:
    continue
  print(number)

```

```python
#Program 2: Password Checker
correct_password = "abc123"
attempts = 0
max_attempts = 3

while attempts < max_attempts:
  password = input("Enter Password: ")
  if password == correct_password:
    print("Access Granted.")
    break

  else:
    attempts +=1
    print("Incorrect Password. Try Again.")
if attempts == max_attempts:
  print("Too Many Attempts. Acces Denied.")





```

```python
#Program 3: Number Filter Display
for number in range(1,11):
  if number == 4:
    continue
  if number == 6 :
    pass
  if number == 8:
    break

  print(number)

```
